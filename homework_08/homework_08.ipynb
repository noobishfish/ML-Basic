## Задание 5

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
np.random.seed(42)
A = np.random.randn(n, n)

U, S, Vt = np.linalg.svd(A)

errors = []
r_values = range(2, n + 1)

for r in r_values:
    U_r = U[:, :r]
    S_r = np.diag(S[:r])
    Vt_r = Vt[:r, :]
    
    A_tilde = U_r @ S_r @ Vt_r
    error = np.linalg.norm(A - A_tilde, ord='fro')
    errors.append(error)

plt.figure(figsize=(10, 6))
plt.plot(r_values, errors, 'b-', linewidth=2, label='Ошибка аппроксимации')
plt.xlabel('Ранг r')
plt.ylabel('Ошибка E(r)')
plt.title(f'Ошибка аппроксимации матрицы {n}×{n} от ранга r')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

Вывод: чем выше ранг матрицы, тем ниже ошибка аппроксимации.
